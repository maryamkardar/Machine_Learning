```python
from scipy import sparse

# Create a random array with a lot of zeros
rnd = np.random.RandomState(seed=123)

X = rnd.uniform(low=0.0, high=1.0, size=(10,5))
print(X)
```

```python
# Set the majority of elements to zero
X[X<0.7] = 0
print(X)
```

```python
# Turn X into a CSR (Compressed_Sparse_Row) matrix
X_csr = sparse.csr_matrix(X)
print(X_csr)
```

```python
# Converting the sparse matrix to a dense array
print(X_csr.toarray())
```

```python
# 1.2 loading digits data
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
digits.keys()
```

```python
n_samples, n_features = digits.data.shape # 1797 images, 64 features for each image
print((n_samples, n_features))  #((.....))
```

```python
print(digits.data[0])  # the data is an array of length 64 (64 features of first image)
print(digits.target_names)  # numbers from 0 to 9
```

```python
# we have two versions of the data array: data and images
print(digits.data.shape)
print(digits.images.shape) # image 8*8 reshaping 64
```

```python
import numpy as np
print(np.all(digits.images.reshape((1797, 64))==digits.data))
```

```python
# set up the figures
fig = plt.figure(figsize=(6,6))  # figure size in inches
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0)

# plot the digits: each image is 8x8 pixels
for i in range(64):
    ax = fig.add_subplot(8, 8, i+1, xticks=[], yticks=[])
    ax.imshow(digits.images[i], cmap=plt.cm.binary, interpolation='nearest')

    #label the image with the target value
    ax.text(0, 7, str(digits.target[i]))
```
